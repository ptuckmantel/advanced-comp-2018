```python
def plot_surface(clf, X, y, n_steps=150,subplot=None, show=False,
                 ylim=None, xlim=None):
    if subplot is None:
        fig = plt.figure()
    else:
        plt.subplot(*subplot)

    if xlim is None:
        xlim = X[:, 0].min(), X[:, 0].max()
    if ylim is None:
        ylim = X[:, 1].min(), X[:, 1].max()
    xx, yy = np.meshgrid(np.linspace(xlim[0], xlim[1], n_steps),
                         np.linspace(ylim[0], ylim[1], n_steps))

    z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
    z=np.array([i[1] for i in z])

    z = z.reshape((xx.shape[0], yy.shape[0]))
    plt.contourf(xx, yy, z, alpha=0.8, cmap=plt.cm.RdBu_r)
    plt.xlim(*xlim)
    plt.ylim(*ylim)

    if show:
        plt.show()
```

```python
print("Decision surface")
plot_surface(model, X,Y_)
plt.scatter(X[:,0], X[:,1], marker='o',c=y, s=50,cmap=plt.cm.RdBu_r)
```

```python
from scipy import ndimage
from PIL import Image



def allequal(iterator):
    iterator = iter(iterator)
    try:
        first = next(iterator)
    except StopIteration:
        return True
    return all(first == rest for rest in iterator)

def add_dim(array):
    y=array[np.newaxis,...]
    return y


def predict_img(input_image, model, predictmodel):
    labels=['a T-shirt/top', 'a Trouser', 'a Pullover', 'a Dress', 'a Coat', 'a Sandal', 'a Shirt', 'a Sneaker', 'a Bag', 'an Ankle boot']
    
    if predictmodel=='nn':
        input_image=add_dim(input_image)
    elif predictmodel=='rf':
        input_image=np.reshape(input_image, 784)
        input_image=add_dim(input_image)
        
    prediction=model.predict(input_image)
    if allequal(prediction[0]):
        label="No idea"
    else:
        label=labels[np.argmax(prediction)]
    
    return label


def id_nn_rf(input_image, nnmodel, rfmodel):
    nnlabel=predict_img(input_image, nnmodel, predictmodel='nn')
    rflabel=predict_img(input_image, rfmodel, predictmodel='rf')
    plt.imshow(input_image)
    plt.show()
    print("NN predicts: %s.\n RF predicts: %s" %(nnlabel, rflabel))
    
    
def prep_image(path):
    image=np.array(Image.open(path))
    image=image/image.max()
    image=image.astype('float32')
    return image
    
```
